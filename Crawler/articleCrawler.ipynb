```python
import requests

url = 'http://bandalmoon.com/api/api_cont_info.php'
res = requests.post(url, data={'category': '1', 'page': '0'})

import json

jsonData = json.loads(res.text)

import re
from IPython.display import HTML


for cont_info in jsonData['cont_info']:
    sub = cont_info['subject']
    popMatch = re.search(r'광고', sub, re.U)
    
    if popMatch != None:
        continue
    
    display(HTML(cont_info['content']))
    break
```
