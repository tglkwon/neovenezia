```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(iris.iloc[:,:-1], iris.iloc[:,-1])
```

```python
plot.precision_recall(iris.iloc[:,-1], rf.predict_proba(iris.iloc[:,:-1]))
```

```python
plot.feature_importances(rf, top_n=3)
```

```python
plot.roc(iris.iloc[:,-1], rf.predict_proba(iris.iloc[:,:-1]))
```

```python
from sklearn.model_selection import learning_curve, validation_curve
```

```python
train_sizes, train_scores, test_scores = learning_curve(
        rf, iris.iloc[:,:-1], iris.iloc[:,-1])
```

```python
plot.learning_curve(train_scores, test_scores, train_sizes)
```

```python
param_range = np.array([1, 10, 100])
param_name = "n_estimators"
train_scores, test_scores = validation_curve(rf, iris.iloc[:,:-1], iris.iloc[:,-1], param_name=param_name, param_range=param_range)
```

```python
plot.validation_curve(train_scores, test_scores, param_range, param_name)
```
